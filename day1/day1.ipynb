```python
import numpy as np
data = np.loadtxt("puzzle1input.dat", dtype=int)
print("Puzzle 1 solution:", data.sum())
```

## Puzzle 2

```python
from collections import deque
i = 0
freq = 0
data_length = len(data)
previous_freqs = deque([0])
while True:
    freq += data[i % data_length]
    if freq in previous_freqs:
        print("Puzzle 2 Solution:", freq)
        break
    else:
        previous_freqs.append(freq)
    if i % 1000 == 0:
        print("i:", i)
        print("freq:", freq)
    i += 1
```
